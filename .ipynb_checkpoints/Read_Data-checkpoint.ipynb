```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("charac_curve/tek0000CH1.csv", skiprows = 20)
print(df.head())

t_0 = np.logical_and((df.iloc[:,0]>=-1e-4),(df.iloc[:,0]<=2e-2+.2e-3))

t = df["TIME"][t_0]
U = df.iloc[:,1][t_0]

plt.plot(t,U)
plt.grid()
plt.show()
```

```python
df = pd.read_excel("PCB_charac.xlsx")
plt.plot(df.iloc[:,0], df.iloc[:,1],"bx-", label='_nolegend_')
plt.plot(2*[32],[0,100],":r", label = "saturation")
plt.plot(2*[217],[0,100],":r")

plt.text(32,-12.3, 32, color="black", ha="center")
plt.text(217,-12.3, 217, color="black", ha="center")
plt.title("Circuit characterization")
plt.xlabel("channel value")
plt.ylabel("duty-cycle [%]")
plt.xlim([0,255])
plt.grid()
#plt.legend(loc="best")
plt.savefig('Circuit_characterization.eps', format='eps', dpi=1000)
plt.show()
```
