```python
# Analyze trajectory of SPA
# Tested with different duty-cycle values controlled with an input ranging from 32(min) - 217 (max)
# between each step there was 300hs delay

df_SPA1 = pd.read_excel("SP2_cropped.xlsx", skiprows = 12, skip_footer = 6) # read and skip useless data
print(df_SPA1.head(3))
print(df_SPA1.tail(3))

t = df_SPA1["t"]
y = df_SPA1["y"]

def to_hs(time): # time format in hr:min:sec:hs, convert to hs
    _, mn, sc, hs = [float(x) for x in str(time).split(':')]
    return mn*60*100 + sc*100 + hs

t_hs = [to_hs(t_elem) for t_elem in t] # list containing the time in hundreds of seconds

i = 0
init_step = 32 # Started with input 32
init_step_time = 90 # first change to step 33 at time : 90hs
LED_CHIP_STEP = [] # Array that will contain the input value (32-217)
# Check for each time element,
# if 300 hs passed, since it is the delay between each step
# if the case increment the input value by 1
for t_hs_elem in t_hs:
    if t_hs_elem>= init_step_time + 300*i:
        i+=1  
    step = init_step + i
    LED_CHIP_STEP.append(step)

    
plt.plot(LED_CHIP_STEP,y*10)
plt.axvline(32, color='r',linestyle='dotted')
plt.axvline(217, color='r',linestyle='dotted')
plt.title("SPA characterization")
plt.xlabel("channel value")
plt.ylabel("displacement [mm]")
plt.xlim([0,255])
plt.grid()
plt.show()
```

```python
from mpl_toolkits.axes_grid1 import host_subplot

host = host_subplot(111)
host.set_xlim(0, 255)
host.set_ylim(-5, 105)

par1 = host.twinx()
par1.set_ylim(10*min(y)-1,1)
par1.set_xlim(0, 255)

host.set_xlabel("channel value")
host.set_ylabel("duty-cycle [%]")
par1.set_ylabel("displacement [mm]")

p1, = host.plot(df_PCB.iloc[:,0], df_PCB.iloc[:,1])
p2, = par1.plot(LED_CHIP_STEP,y*10)
par1.axvline(32, color='r',linestyle='dotted')
par1.axvline(217, color='r',linestyle='dotted')


host.axis["left"].label.set_color(p1.get_color())
par1.axis["right"].label.set_color(p2.get_color())

plt.grid()
plt.draw()
plt.show()
```
